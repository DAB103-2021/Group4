### Data Transformation

##### Converting float data type interger to  data type

```python
calo.info()
```

```python
calo["Saturated Fat"] = calo["Saturated Fat"].astype(np.int64)
calo["Trans Fat"] = calo["Trans Fat"].astype(np.int64)
calo["Fiber"] = calo["Fiber"].astype(np.int64)
calo["Protein"] = calo["Protein"].astype(np.int64)
```

```python
calo.info()
```

##### Converting grams into mili grams except Cholesterol and Sodium

```python
calo["Calories"]=calo["Calories"] * 1000
calo["Calorie Fat"]=calo["Calorie Fat"] * 1000
calo["Total Fat"]=calo["Total Fat"] * 1000
calo["Saturated Fat"]=calo["Saturated Fat"] * 1000
calo["Trans Fat"]=calo["Trans Fat"] * 1000
calo["Total Carbohydrates"]=calo["Total Carbohydrates"] * 1000
calo["Fiber"]=calo["Fiber"] * 1000
calo["Sugar"]=calo["Sugar"] * 1000
calo["Protein"]=calo["Protein"] * 1000

calo.head()
```

##### Converting Vitamin A, C and Calcium from percentage(%) to miligrams(mg) for age children and adult

```python
calo["Vitamin A (18+)"]=20*calo["Vitamin A"]/100*1000
calo["Vitamin C (18+)"]=20*calo["Vitamin C"]/100*1000
calo["Calcium (18+)"]=20*calo["Calcium"]/100*1000

calo["Vitamin A (18-)"]=12*calo["Vitamin A"]/100*1000
calo["Vitamin C (18-)"]=12*calo["Vitamin C"]/100*1000
calo["Calcium (18-)"]=12*calo["Calcium"]/100*1000

calo.head()
```
